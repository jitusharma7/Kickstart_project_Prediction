```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import fnmatch
import matplotlib.pyplot as plt
```

```python
## Checking for missing values:
print(kick_data_1.isnull().sum())
kick_data_1 = kick_data_1.dropna()
print(kick_data_1.isnull().sum())

print(kick_data_1.shape)
```

```python
#Implementing k - Nearest Neighbour Algorithm:
```

```python
## Converting categorical data into numerical data:
kick_data_1.category = pd.Categorical(kick_data_1.category).codes
kick_data_1.State = pd.Categorical(kick_data_1.State).codes
kick_data_1.City = pd.Categorical(kick_data_1.City).codes
kick_data_1.subcategory = pd.Categorical(kick_data_1.subcategory).codes
kick_data_1.status = pd.Categorical(kick_data_1.status).codes
```

```python
kick_data_1.category = pd.Categorical(kick_data_1.category).codes
```

```python
kick_data_1.State = pd.Categorical(kick_data_1.State).codes
```

```python
kick_data_1.City = pd.Categorical(kick_data_1.City).codes
```

```python
kick_data_1.subcategory = pd.Categorical(kick_data_1.subcategory).codes
```

```python
kick_data_1.status = pd.Categorical(kick_data_1.status).codes
```

```python
## Dividing into training and test data set
y1 = kick_data_1['status']
kick_data_1_2 = kick_data_1[['category','subcategory','City','State','goal','levels','duration','updates']]
x_trn, x_tst, y_trn, y_tst = train_test_split(kick_data_1_2,y1, test_size = 0.3)
```

```python
## k - Nearest Neighbour Model:
knn = KNeighborsClassifier(n_neighbors=7)
```

```python
## Train the model using training data set
knn.fit(x_trn, y_trn)
```

```python
y_pred = knn.predict(x_tst)
```

```python
## Model Accuracy and F1-Score
cfm = metrics.confusion_matrix(y_tst,y_pred)
acc = metrics.accuracy_score(y_tst, y_pred)
f1_score = metrics.f1_score(y_tst,y_pred)
report = metrics.classification_report(y_tst,y_pred)
print("Confusion  Matrix", cfm)
print("Accuracy:",acc*100)
print("Error:", (1 - acc)*100)
print("F1 Score:",f1_score)
print("Classfication Report:",report)
```
